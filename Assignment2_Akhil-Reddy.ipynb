## 1.  How many passengers were there in each class?

```python
filename = "titanic3.csv"
df = pd.read_csv( filename, header=0 )
df_pclass = df['pclass'].value_counts(ascending=True).reset_index()
# set column names
df_pclass.columns = ['pclass', 'passenger_sum']
df_pclass.head()
```

```python
# import bokeh 
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.models import HoverTool

# tell bokeh to show the figures in the notebook
from bokeh.io import output_notebook
output_notebook()
```

```python
# Create input data for bokeh from the dataframe
source = ColumnDataSource( df_pclass )
```

```python
#Hover
hover = HoverTool(
    tooltips=[("Passenger Class", '@pclass'),
              ("Total Count", '@passenger_sum')])

p = figure( title="Passengers Count in each Class", plot_width=500, x_axis_label='Passenger Class',
           y_axis_label='Count of Passengers', tools=[hover],
           x_axis_type='linear')
#p.hbar( source=source, y='pclass', height=0.1, right='count')
p.vbar( source=source, x='pclass', width=0.5, bottom=0, top='passenger_sum')
show(p)
```

## 2. Display top 20 cabin's according to the number of passengers staying?

```python
filename = "titanic3.csv"
df = pd.read_csv( filename, header=0 )

df_cabin = df['cabin'].value_counts(ascending=False).reset_index()
df_cabin = df_cabin.reindex(index=df_cabin.index[::]).head(20)
df_cabin.columns=['cabin','count']
df_cabin
```

```python
# import bokeh 
from bokeh.models.renderers import GlyphRenderer
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.models import HoverTool

# tell bokeh to show the figures in the notebook
from bokeh.io import output_notebook
output_notebook()
source= ColumnDataSource( df_cabin )

hover = HoverTool(
    tooltips=[("Cabin Class", '@cabin'),
              ("Total Count", '@count')])

p = figure(y_range=df_cabin['cabin'].tolist(), title="Passengers Per Cabin", plot_width=800, plot_height=600,
        x_axis_label='Count',
          y_axis_label='Cabin Class', tools=[hover])
p.hbar(source=source, y='cabin', height=0.3, right='count')
show(p)
```

## 3. Find the total Number of Siblings/Spouses Aboard and Number of Parents/Children Aboard according to the pasenger class?

```python
filename = "titanic3.csv"
df = pd.read_csv( filename, header=0 )

total_sum = df[['pclass','sibsp','parch']].groupby('pclass', as_index=False).sum()
total_sum
```

```python
# import bokeh 
from bokeh.models.renderers import GlyphRenderer
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.models import HoverTool

# tell bokeh to show the figures in the notebook
from bokeh.io import output_notebook
output_notebook()
```

```python
source = ColumnDataSource( total_sum )
```

```python
#Hover
hover = HoverTool(
    tooltips=[("Passenger Class", '@pclass'),
              ("Total Count of sibsp", '@sibsp')])
p = figure( title="Passengers", plot_width=500, x_axis_label='Passenger Class',
          y_axis_label='Count of Siblings', tools=[hover],
   x_axis_type='linear')

p.vbar( source=source, x='pclass', width=0.5, bottom=0, top='sibsp')     

# add a line renderer
show(p)
```

```python
#Hover
hover_line = HoverTool(
    tooltips=[("Passenger Class", '@pclass'),
              ("Total Count of parch", '@parch')])

p1 = figure(plot_width=500, plot_height=500, tools=[hover_line] , x_axis_type='linear')
# add a line renderer
p1.circle(source=source, x='pclass', y='parch', line_width=5, color='red')
show(p1)
```

## 4. Number of People survived in each  pClass?

```python
filename = "titanic3.csv"
df = pd.read_csv( filename, header=0 )
survived_sum = df[['pclass','survived']].groupby('pclass', as_index=False).sum()
survived_sum
```

```python
# import bokeh 
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.models import HoverTool

# tell bokeh to show the figures in the notebook
from bokeh.io import output_notebook
output_notebook()
```

```python
# Create input data for bokeh from the dataframe
source = ColumnDataSource( survived_sum )
```

```python
#Hover
hover = HoverTool(
    tooltips=[("Passenger Class", '@pclass'),
              ("Total Count", '@survived')])

p = figure( title="Passengers Survived in each Class", plot_width=500, x_axis_label='Passenger Class',
           y_axis_label='Count of Passengers', tools=[hover],
           x_axis_type='linear')
p.vbar( source=source, x='pclass', width=0.5, bottom=0, top='survived')
show(p)
```

## 5. Count the number of passengers who embarked at different locations?

```python
embarked_count = df['embarked'].value_counts().reset_index()
embarked_count.columns = ['embarked', 'emb_count']
embarked_count.head()
```

```python
# import bokeh 
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.models import HoverTool
from bokeh.models import Span

# tell bokeh to show the figures in the notebook
from bokeh.io import output_notebook
output_notebook()
```

```python
embarked_value = embarked_count['embarked'].tolist()
count_value = embarked_count['emb_count'].tolist()

p = figure(y_range=embarked_value)
p.circle(count_value, embarked_value, size=15, fill_color="red", line_color="black", line_width=3)
show(p)
```

## 6. Group passengers by age and tell what is the average age of passengers travelling?

```python
age_count = df['age'].value_counts().reset_index()
age_count.columns = ['age', 'count']
age_count.head()
```

```python
# import bokeh 
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.models import HoverTool
from bokeh.models import Span

# tell bokeh to show the figures in the notebook
from bokeh.io import output_notebook
output_notebook()
```

```python
# Create input data for bokeh from the dataframe
source = ColumnDataSource( age_count )
```

```python
#Hover
hover_count = HoverTool(
    tooltips=[("Passenger count", '@count')])

p = figure( title="Average Age of the Passengers", plot_width=500, x_axis_label='Age',
           y_axis_label='Count', tools=[hover_count])
#p.hbar( source=source, y='pclass', height=0.1, right='count')
p.vbar( source=source, x='age', width=0.5, bottom=0, top='count')
vline = Span(location=df['age'].mean(),dimension='height',line_dash='dashed', line_width=3, 
               line_color='red')
p.renderers.extend([vline])
show(p)
```
